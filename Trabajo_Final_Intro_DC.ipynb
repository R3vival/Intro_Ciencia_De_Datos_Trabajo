### ¿Qué es y qué nos dice el `df.sample()`?

`df.sample()` es una función que selecciona una fila completamente al azar del dataset. La cual cada vez que la ejecutamos, nos mostrará una fila diferente.

**¿Para qué sirve?**
Esto es útil porque, a diferencia de `df.head()` que siempre muestra las primeras filas, aquí obtenemos un registro aleatorio que podría venir de cualquier parte del dataset.

**¿Qué podemos verificar con esta fila aleatoria?**

| Verificación | ¿Qué buscamos? |
| :--- | :--- |
| **Tipos de datos coherentes** | Que los números sean números y el texto sea texto |
| **Valores dentro de rangos lógicos** | Que no haya datos absurdos como una edad de 500 años o una duración de sueño de -3 horas |
| **Variables binarias (0 o 1)** | Que columnas como `exercise_day`, `felt_rested`, `sleep_aid_used` y `shift_work` solo contengan 0 (no) o 1 (sí) |
| **Variables categóricas válidas** | Que columnas como `chronotype`, `season` o `sleep_disorder_risk` tengan categorías que tengan sentido |
| **Sin valores vacíos** | Que la fila tenga todos sus 31 campos completos, sin datos faltantes |

**En resumen:** Esta revisión rápida nos da confianza de que los datos están bien formados antes de hacer un análisis. Si algo raro apareciera aquí (un texto donde debería haber un número, un valor imposible), sería una señal de alerta para limpiar los datos.

## Paso 1: Calidad de los Datos
Buscamos celdas vacías (nulos) y filas repetidas (duplicados).

```python
print("Valores nulos por columna:")
display(df.isnull().sum())

print("\nCantidad de registros duplicados:")
display(df.duplicated().sum())
```

```python
#Mapeo de datos faltantes

plt.figure(figsize=(10, 7))
sns.heatmap(df.isna(), cbar=False)
plt.title("Mapa de valores faltantes")
plt.tight_layout()
plt.show()
```

**Valores nulos (datos vacíos):** Las 31 columnas tienen 0 nulos. Esto significa que no hay celdas vacias, cada persona tiene todos sus datos completos.

**Filas duplicadas:** Hay 0 duplicados. Ninguna fila se repite exactamente igual. Esto es importante porque si hubiera filas repetidas, estaríamos "contando doble" a una persona y eso podria alterar nuestros resultados.

Por ende, en el mapeo de datos faltantes, no observamos novedades, puesto que el dataset se encuentra a plenitud para un analisis sin errores del muestreo de datos.3

## Paso 2: Análisis de Variables Numéricas
Exploramos la distribución de los datos y buscamos valores atípicos (outliers) mediante Boxplots.

```python
# Configuramos el estilo de los gráficos
sns.set_theme(style="whitegrid")

# Creamos gráficos para variables clave
fig, axes = plt.subplots(figsize=(15, 10))

sns.boxplot(x=df['sleep_duration_hrs'], ax=axes, color='lightblue')
axes.set_title('Distribución: Horas de Sueño')

plt.tight_layout()
plt.show()
```

**Horas de Sueño**

La caja va más o menos de las 5.5 a 7 horas. Esto significa que la mayoría de la gente del analisis duerme en ese rango de horas.

Por otro lado, como podemos observad hay outliers a la derecha, son pocas personas que duermen mucho más que el resto.

**Conclusión:** la mayoría de pesonas duermen entre 5 y 7 horas, que sigue siendo poco ya que lo recomendado es por lo menos entre 7 y 8 horas. Los que duermen 9-10 son excepciones raras.

```python
fig, axes = plt.subplots(figsize=(15, 10))

sns.boxplot(x=df['stress_score'], ax=axes, color='lightcoral')
axes.set_title('Distribución: Nivel de Estrés')
```

```python
valores_unicos = df["stress_score"].unique()
print(valores_unicos)
```

**Nivel de Estrés**

Para el nivel de estres, nos basamos en una escala que va del 1 al 10, donde de 1 a 3.9 es un nivel de estre es bajo, de 4 a 6.9 es un nivel medio y de 7 a 10 es un nivel de estres alto.

La caja va de 4 a 7 aproximadamente, con la mediana cerca de 5-6. Hay outliers en ambos extremos.

**Conclusión:** la mayoría tiene estrés relativamente moderado. Los extremos son personas muy relajadas o muy estresadas, son casos raros.
