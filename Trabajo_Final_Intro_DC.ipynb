# **Nombre del estudio:** Salud del Sueño y Rendimiento Diario

Este conjunto de datos contiene 100,000 registros de pacientes sintéticos diseñados para un análisis integral de la salud del sueño, la predicción del riesgo de trastornos y el modelado del rendimiento cognitivo. Cada registro captura una instantánea diaria completa: mediciones de la arquitectura del sueño, comportamientos de estilo de vida, estado psicológico y resultados cognitivos del día siguiente.
