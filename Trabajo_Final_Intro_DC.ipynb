### ¿Qué es y qué nos dice el `df.sample()`?

`df.sample()` es una función que selecciona una fila completamente al azar del dataset. La cual cada vez que la ejecutamos, nos mostrará una fila diferente.

**¿Para qué sirve?**
Esto es útil porque, a diferencia de `df.head()` que siempre muestra las primeras filas, aquí obtenemos un registro aleatorio que podría venir de cualquier parte del dataset.

**¿Qué podemos verificar con esta fila aleatoria?**

| Verificación | ¿Qué buscamos? |
| :--- | :--- |
| **Tipos de datos coherentes** | Que los números sean números y el texto sea texto |
| **Valores dentro de rangos lógicos** | Que no haya datos absurdos como una edad de 500 años o una duración de sueño de -3 horas |
| **Variables binarias (0 o 1)** | Que columnas como `exercise_day`, `felt_rested`, `sleep_aid_used` y `shift_work` solo contengan 0 (no) o 1 (sí) |
| **Variables categóricas válidas** | Que columnas como `chronotype`, `season` o `sleep_disorder_risk` tengan categorías que tengan sentido |
| **Sin valores vacíos** | Que la fila tenga todos sus 31 campos completos, sin datos faltantes |

**En resumen:** Esta revisión rápida (conocida como Sanity Check visual) nos da una primera impresión sobre la estructura del dataset. Sin embargo, dado que una muestra de 5 filas no es estadísticamente representativa para 100.000 registros, utilizaremos herramientas programáticas (como df.describe() y df.isnull()) en los siguientes pasos para garantizar matemáticamente que estas condiciones se cumplen en el 100% de los datos.

Basándonos en la distribución matemática y en la lógica del negocio, se decide aplicar una técnica de discretización (Binning) para crear una nueva variable categórica llamada `stress_category`. Esta agrupará a los pacientes en tres cohortes de riesgo: Estrés Bajo (1 a 3.9), Medio (4 a 6.9) y Alto (7 a 10), lo cual mejorará significativamente el rendimiento predictivo de nuestro futuro modelo de Machine Learning.

```python
# ---------------------------------------------------------
# Creacion de variable Categorica
# ---------------------------------------------------------
# Aplicamos la regla de negocio definida en el análisis exploratorio
limites = [0, 3.9, 6.9, 10]  # El 0 asegura que el 1 quede incluido
etiquetas = ['Bajo', 'Medio', 'Alto']

# Creamos la nueva variable categórica ordinal
df['stress_category'] = pd.cut(df['stress_score'], bins=limites, labels=etiquetas)

# Verificamos cuántas personas quedaron en cada grupo
print(df['stress_category'].value_counts())
```

```python
# ==========================================
# 1. ANÁLISIS VISUAL: Énfasis en la Métrica Clave (Severe)
# ==========================================
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculamos los porcentajes
crosstab_perc = pd.crosstab(df['occupation'], df['sleep_disorder_risk'], normalize='index') * 100

# Ordenamos
if 'Severe' in crosstab_perc.columns:
    crosstab_perc = crosstab_perc.sort_values(by='Severe', ascending=False)

# Graficamos
ax = crosstab_perc.plot(kind='barh', stacked=True, figsize=(14, 8), colormap='RdYlGn_r')

# TRUCO SENIOR: Damos un poco de margen extra al final (hasta 110%) para que quepa el texto
ax.set_xlim(0, 110)

# --- EL TOQUE SENIOR: Data Storytelling ---
for c in ax.containers:
    # Identificamos a qué categoría (nivel de riesgo) pertenece este bloque de color
    categoria = c.get_label()

    for v in c:
        width = v.get_width()

        if width > 0:
            texto = f'{width:.1f}%'

            # Si es el riesgo SEVERE, lo ponemos AFUERA, al final de la barra (x=101)
            if categoria == 'Severe':
                x = 101  # Justo pasando el límite del 100%
                y = v.get_y() + (v.get_height() / 2) # Centro vertical

                # Lo pintamos más grande y en color rojo oscuro para que llame la atención
                ax.text(x, y, texto, ha='left', va='center',
                        fontsize=11, weight='bold', color='darkred')

            # Para los demás riesgos, los dejamos centrados y discretos dentro de su color
            else:
                x = v.get_x() + (width / 2)
                y = v.get_y() + v.get_height()
                ax.text(x, y, texto, ha='center', va='bottom', fontsize=9, weight='bold', color='black')

plt.title('Proporción de Riesgo de Trastorno del Sueño por Ocupación', fontsize=14, pad=20)
plt.xlabel('Porcentaje de Pacientes (%)')
plt.ylabel('Ocupación')

# Movemos la leyenda afuera
plt.legend(title='Nivel de Riesgo', bbox_to_anchor=(1.05, 1), loc='upper left')

# Eliminamos las líneas negras superior y derecha para un look de consultoría
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

La visualización revela que los Enfermeros y los Ingenieros de Software son los grupos con mayor vulnerabilidad clínica. Casi el 20% de los enfermeros presentan un riesgo Severo, la tasa más alta de todo el dataset. Esto correlaciona con la literatura médica sobre el impacto del trabajo por turnos y la alta carga cognitiva en la higiene del sueño.

Segmento de Bajo Riesgo: Por el contrario, los Abogados y las Amas de casa presentan los perfiles más saludables, con más del 85% de su población en la categoría 'Healthy'. Es notable que, a pesar de las responsabilidades asociadas, su estructura de riesgo severo es casi inexistente (0.3% - 0.4%).

Paradoja del Estudiante: Aunque en la gráfica anterior vimos que los estudiantes tenían el estrés más alto, esta gráfica muestra que su riesgo Severo es relativamente bajo (5.4%) comparado con los enfermeros. Esto sugiere una resiliencia biológica por edad o que su estrés no se traduce inmediatamente en una patología severa del sueño, sino quizás en una calidad pobre.

**Los más estresados:**
Abogados lideran por lejos el ranking de estrés. Enfermeros y Doctores les siguen de cerca.

**Estrés medio:**
Conductores, Managers, Estudiantes, Profesores y Ventas se agrupan entre 5.5 y 6.5. Son profesiones con estrés pero más manejable.

**Los menos estresados:**
Freelancers y Amas de casa tienen niveles moderados. Jubilados son por lejos los menos estresados, lo cual es completamente lógico.

Las profesiones del área de salud (enfermeros, doctores) están entre las más estresadas, lo que es relevante porque el estrés es uno de los factores que más afecta la calidad del sueño.

Por consiguiente, si sos abogado, enfermero o doctor, los datos dicen que probablemente estés mucho más estresado que alguien jubilado o freelancer. Ahora bien, con los analisis previamente realizados y teniendo encuenta que el estres es uno de los factores que mas afecta el sueño, estas profesiones podrían ser las que peor duermen, segun este muestreo que data de 100.000 personas.
